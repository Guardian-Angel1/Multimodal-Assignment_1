# **Indi_Weather Analysis: Understanding Climate Trends in Indian Cities**

## **Objective**
This analysis explores historical weather trends in major Indian cities using a dataset containing temperature, humidity, and wind speed data. We use various **visualizations** to extract insights that can be useful for **urban planning, climate research, and renewable energy projects**.

---

## **1. Temperature Trends Across Cities**
### **Goal:** 
Plot a line graph to show how temperatures change over time in different cities.

---

## **2. Average Temperature Across Cities (Bar Chart)**
### **Goal:** 
Compare the **average temperature** of different cities over the last 30 days.

---

## **3. Humidity Levels Across Cities (Box Plot)**
### **Goal:** 
Visualize the **humidity distribution** for different cities.

---

## **4. Wind Speed Variations Across Cities (Violin Plot)**
### **Goal:** 
Analyze wind speed patterns to identify potential locations for **wind energy projects**.

---

## **5. Correlation Heatmap (Temperature, Humidity, Wind Speed)**
### **Goal:** 
Identify relationships between weather parameters like **temperature, humidity, and wind speed**.

---

## **Insights from the Visualizations**
- **Temperature trends** help track seasonal variations in different cities.
- **Bar charts** highlight cities with extreme temperature variations.
- **Box plots** and **violin plots** reveal humidity and wind speed fluctuations, aiding urban planning.
- **Heatmaps** show correlations between weather variables, useful for climate modeling.

### **Future Applications**
- Identifying **heatwave-prone cities** to improve urban cooling strategies.

- Selecting **optimal locations for renewable energy projects** (solar & wind energy).

```python
import pandas as pd
df = pd.read_csv("Indi_Weather_Trends.csv")
print(df.head())
```

```python
avg_temp = df.groupby("city")["temperature"].mean().sort_values(ascending=False)

print("Hottest City:", avg_temp.idxmax(), "with Avg Temp:", avg_temp.max())
print("Coldest City:", avg_temp.idxmin(), "with Avg Temp:", avg_temp.min())
```

```python
high_humidity_cities = df.groupby("city")["humidity"].mean()
high_humidity_cities = high_humidity_cities[high_humidity_cities > 80]

print("Cities with high humidity:\n", high_humidity_cities)
```

```python
windy_cities = df.groupby("city")["wind_speed"].mean()
windy_cities = windy_cities[windy_cities > 10]

print("Potential cities for wind energy:\n", windy_cities)
```

```python
import matplotlib.pyplot as plt

city_data = df[df["city"] == "Mumbai"]

city_data["timestamp"] = pd.to_datetime(city_data["timestamp"])

plt.figure(figsize=(10, 5))
plt.plot(city_data["timestamp"], city_data["temperature"], label="Temperature (C)")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.title("Temperature Trend in Mumbai")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

selected_cities = ["Mumbai", "Delhi", "Bangalore", "Chennai", "Kolkata"]

df["timestamp"] = pd.to_datetime(df["timestamp"])

plt.figure(figsize=(12, 6))

for city in selected_cities:
    city_data = df[df["city"] == city]
    plt.plot(city_data["timestamp"], city_data["temperature"], label=city)

plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.title("Temperature Trends Across Major Cities")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

avg_temp_city = df.groupby("city")["temperature"].mean().sort_values()

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_temp_city.values, y=avg_temp_city.index, palette="coolwarm")
plt.xlabel("Average Temperature (C)")
plt.ylabel("City")
plt.title("Average Temperature in Indian Cities (Last 30 Days)")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df["city"], y=df["humidity"], palette="viridis")
plt.xlabel("City")
plt.ylabel("Humidity (%)")
plt.title("Humidity Distribution Across Cities")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x=df["city"], y=df["wind_speed"], palette="Blues")
plt.xlabel("City")
plt.ylabel("Wind Speed (m/s)")
plt.title("Wind Speed Variations Across Cities")
plt.xticks(rotation=90)
plt.show()
```

```python
import numpy as np

corr_matrix = df[["temperature", "humidity", "wind_speed"]].corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Weather Parameters")
plt.show()
```
